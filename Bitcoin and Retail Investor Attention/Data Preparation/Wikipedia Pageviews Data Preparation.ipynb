# Preparing Wikipedia Pageviews Data

The data is taken from the Wikipedia Pageviews Analysis tool (https://tools.wmflabs.org/pageviews/?project=en.wikipedia.org&platform=all-access&agent=user&start=2015-07-01&end=2018-04-14&pages=Bitcoin) for the anglophone article "Bitcoin".
It represents the daily pageviews on all platforms from 07/01/2015 - 04/14/2018.

Import used libraries

```python
import numpy as np
import pandas as pd
```

## Import the CSV file (wikipedia_pageviews_bitcoin.csv) containing the pageview data

```python
wikipedia_pageviews_raw = pd.DataFrame.from_csv("wikipedia_pageviews_bitcoin.csv")
wikipedia_pageviews_finished = wikipedia_pageviews_raw.reset_index()
wikipedia_pageviews_finished = wikipedia_pageviews_finished.rename(columns={"Bitcoin" : "Pageviews"})
wikipedia_pageviews_finished = wikipedia_pageviews_finished.drop(["Date"], axis=1)
```
