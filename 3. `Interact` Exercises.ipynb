### Reverse some text

Write a function that takes text as an input and reverses the text, and uses `interact` or `interactive` to generate the input field. 

```python
@interact(text='Insert your text here!')

def f(text):
    return text[::-1]
```

### Make a plot

Pick a function of interest to you that depends on a couple of parameters and use `interact` to generate controls for the parameters. 

If you don't have a favorite function, try graphing $f(x) = \sin(k x + p)$ over the $0 < x < 4\pi$, with $0.5 \le k \le 2$ and $0 \le p < 2\pi$.

Use either the `interact` function or the `@interact` decorator.

```python
@interact(k=(0.5, 2), p=(0, 2*np.pi))

def g(k=0, p=0):
    fig, ax = plt.subplots()
    x = np.linspace(0, 4*np.pi, 1000)
    ax.plot(x, np.sin(k*x + p))
    
```

### Choose two widgets and link their values

Choose any two widgets from the widget overview, link their values using `widgets.link`, and display both widgets (the `display` function can take more than one argument).

Note: for all widgets except the section widgets below you should link the `value` of the widgets. For the selection widgets, link the `index` instead.

Selection widgets:

+ `Dropdown`
+ `RadioButtons`
+ `Select`
+ `SelectionSlider`
+ `SelectionRangeSlider`
+ `ToggleButtons`
+ `SelectMultiple`

```python
widget_1 = widgets.RadioButtons(
    options=['Icecream', 'Burger', 'Pizza'],
    value='Icecream',
    description='Dinner')
```

```python
widget_2 = widgets.Dropdown(
    options=['Icecream', 'Burger', 'Pizza'],
    value='Icecream',
    description='Lunch')
```

```python
widgets.link((widget_1, 'value'), (widget_2, 'value'));
```

```python
display(widget_1, widget_2)
```

### Boxes

Create any three widgets, place them in a `VBox`, and display it.

```python
widget_1 = widgets.FloatSlider(
    value=1.5,
    min=-5,
    max=15.7,
    description='Widget 1:')
```

```python
widget_2 = widgets.ToggleButton(
    value=False,
    description='Widget 2:')
```

```python
widget_3 = widgets.SelectionSlider(
    options=['Cool', 'Hot', 'Cold', 'Rain'],
    description='Widget 3:')
```

```python
widgets.VBox([widget_1, widget_2, widget_3])
```
